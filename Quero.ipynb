```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from tqdm import tqdm
```

```python
cursos = ['marketing', 'comunicacao-social', 'engenharia-ambiental', 'gestao-ambiental',
          'psicologia', 'ciencia-da-computacao', 'tecnologia-da-informacao', 'psicanalise',
          'redes-de-computadores', 'pedagogia', 'analise-e-desenvolvimento-de-sistemas', 'medicina', 'historia']
cursos_links = []
grade_curricular = []
for i in cursos:
  link = "https://querobolsa.com.br/cursos-e-faculdades/" + i
  link_grade = "https://querobolsa.com.br/cursos-e-faculdades/" + i + "/grade-curricular"
  cursos_links.append(link)
  grade_curricular.append(link_grade)


print(cursos_links)
print(grade_curricular)
```

```python
def get_description_curso(content):
  description = []
  for i in content.find_all("div", {"class": "z-dynamic-content"}):
    description.append(i.text)
  return description
```

```python
def get_materias(content):
  materias = []
  for i in content.find_all("li", {"class": "curriculum__disciplines-item"}):
    materias.append(i.text.strip())
  return materias
```

```python
html_desc = requests.get('https://querobolsa.com.br/cursos-e-faculdades/marketing').text
curso = BeautifulSoup(html_desc, "html")
print(curso)
teste = get_description_curso(curso)
print(teste)
```

```python
legend = ['Nome Curso', 'Descrição', 'Materias', 'Tipo', 'Duração (Semestres)']
```

```python
cursos_final = pd.DataFrame(columns=legend)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
tipo = ['grad', 'grad', 'grad', 'grad ou tec', 'grad', 'grad', 'tec', 'curso livre', 'grad ou tec', 'grad', 'grad', 'grad', 'lic']
duracao = [8,8,10,6,10,8,4,4,6,8,'4-6',12,8]
```

```python
psico = ['Introdução à Psicanálise, Individuação,Metapsicologia 1,Metapsicologia 2,Psicanálise e Sociedade,Metapsicologia 3,Filosofia e Psicanálise,Neuroses 1,Ética na Psicanálise,Perversões,Neuroses 2,Psicoses,Sexologia Aplicada,Fundamentos da Técnica Psicanalítica 1,Psicanálise para Crianças,Psicopatologia Aplicada 1,Psicanálise na Adolescência,Fenômenos Psicossomáticos,Fundamentos da Técnica Psicanalítica,Tópicos Especiais: Emergências. Perdas e lutos,Psicopatologia Aplicada 2,Transtornos Alimentares,Psicanálise na Dependência Química,Teorias Pós Freudianas,Semiologia Psicanalítica,Metapsicologia 4,Psicanálise Contemporânea 1,Metapsicologia 5,Psicanálise Contemporânea 2']
```

```python
aid = ['Análise de Sistemas Orientada a Objetos,Atividades Complementares,Banco de Dados,Comunicação Aplicada,Desenvolvimento de Software para Internet,Desenvolvimento Sustentável,Direitos Humanos (Optativa),Economia e Mercado,Educação Ambiental (Optativa),Empreendedorismo,Engenharia de Software I,Engenharia de Software II,Estatística,Estudos Disciplinares,Ética e Legislação Profissional,Fundamentos de Redes de Dados e Comunicação,Fundamentos de Sistemas Operacionais,Gerenciamento de Projetos de Software,Gestão da Qualidade,Gestão Estratégica de Recursos Humanos,Língua Brasileira de Sinais - LIBRAS - (Optativa),Linguagem e Técnicas de Programação,Lógica,Matemática para Computação,Metodologia Científica,Organização de Computadores,Princípios de Sistemas de Informação,Programação Orientada a Objetos I,Programação Orientada a Objetos II,Projeto de Interface com o Usuário,Projeto de Sistemas Orientado a Objetos,Projeto Integrado Multidisciplinar,Relações Étnico-Raciais e Afrodescendência (Optativa),Tópicos Especiais de Programação Orientada a Objetos']
```

```python
for i in tqdm(range(len(cursos))):
  html_desc = requests.get(cursos_links[i]).text
  curso = BeautifulSoup(html_desc, "html")
  html_mat = requests.get(grade_curricular[i]).text
  materias = BeautifulSoup(html_mat, "html")
  curso_name = cursos[i]
  desc = get_description_curso(curso)
  mats = get_materias(materias)
  if i == 7:
    insert = [curso_name, desc, psico[0], tipo[i], duracao[i]]
    cursos_final.loc[i] = insert
  elif i == 10:
    insert = [curso_name, desc, aid[0], tipo[i], duracao[i]]
    cursos_final.loc[i] = insert
  else:
    insert = [curso_name, desc, mats, tipo[i], duracao[i]]
    cursos_final.loc[i] = insert
```

```python
prof_nomes = ['devops', 'professor', 'cientista-de-dados', 'psicologo-comportamental',
        'analista-ambiental', 'analista-de-marketing-digital']
prof_links = []
for i in prof_nomes:
  link = "https://querobolsa.com.br/carreiras-e-profissoes/" + i
  prof_links.append(link)

print(prof_links)
```

```python
def get_description_prof(content):
  description = []
  for i in content.find_all("div", {"class": "z-dynamic-content"}):
    description.append(i.text)
  return description
```

```python
legend2 = ['Nome Curso', 'Descrição', 'Cursos']
```

```python
cursos_prof = ['ciencia-da-computacao, analise-e-desenvolvimento-de-sistemas',
               'historia, pedagogia', 'ciencia-da-computacao', 'psicologia, medicina, psicanalise',
               'engenharia-ambiental, gestao-ambiental', 'marketing, comunicacao-social']
```

```python
prof_final = pd.DataFrame(columns=legend2)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
for i in tqdm(range(len(cursos_prof))):
  html_desc = requests.get(prof_links[i]).text
  profissao = BeautifulSoup(html_desc, "html")
  prof_name = prof_nomes[i]
  desc = get_description_prof(profissao)
  insert = [prof_name, desc, cursos_prof[i]]
  prof_final.loc[i] = insert
```
